### Detection of Outliers - Unemployment Rate

Using Z score for calculating outliers.

Z Score = (X − μ )/σ

where X = data point,
μ = mean,
σ = Standard Deviation

```python
# python function to calculate outliers

# empty list which will contain outliers
outliers = []

def detect_outliers(dataset_column):
    max_std = 3
    mean = np.mean(dataset_column)
    std = np.std(dataset_column)
    
    for i in dataset_column:
        z_score = (i - mean)/std
        if np.abs(z_score) > max_std:
            outliers.append(i)
    return outliers
```

```python
# changing the names of the column and removing extra spaces
col_names = df.columns.tolist()
col_names
```

```python
df.set_axis(['Region', 'Date', 'Frequency', 'Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)', 'Area' ], axis=1, inplace=True)
df.head()
```

```python
col_names = df.columns.tolist()
col_names
```

The extra spaces have been removed. Let's calculate the outliers. 

```python
unemploy_data = df['Estimated Unemployment Rate (%)']
labor_part_data = df['Estimated Labour Participation Rate (%)']
unemploy_outlier = detect_outliers(unemploy_data)
print(unemploy_outlier)
```

### Interquartile Range - Unemployment Rate


```python
df.describe()
```

```python
# sort the data
sorted(unemploy_data)
```

```python
quartile1, quartile3 = np.percentile(unemploy_data,[25,75])
quartile1 = round(quartile1, 2)
quartile3 = round(quartile3, 2)
print('25th Quartile:',quartile1,'| 75th Quartile:',quartile3)
```

```python
# calculating iqr
iqr_value = quartile3 - quartile1
print(iqr_value)
```

```python
# Find lower bound & upper bound value
lower_bound_val = quartile1 - (1.5 * iqr_value)
upper_bound_val = quartile3 + (1.5 * iqr_value)
lower_bound_val = round(lower_bound_val, 2)
upper_bound_val = round(upper_bound_val, 2)
```

```python
print('Upper Bound:', upper_bound_val,'| Lower Bound:', lower_bound_val)
```
